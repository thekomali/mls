#### NOTATIONS


| Notation | Description | code |
|---|---|---|
| $x$ | training data - features | x_train |
| $y$ | training data - targets | y_train |
| $x^{(i)}$, $y^{(i)}$ | $i_{th}$ training data | `x_i`, `y_i` |
| m | number of training data | `m` |
| w | parameter - weight | `w` |
| b | parameter - bias | `b` |
| $f_{w,b}(x^{(i)})$ | model being evalueated at $x^{(i)}$ parameterized by $w,b$: $f_{w,b}(x^{(i)}) = wx^{(i)}+b$ | `f_wb` |


#### Model Function
the model function for linear regression (which is a function that maps from `x` to `y`) is represented as 

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$

The formula above is how you can represent straight lines - different values of $w$ and $b$ give you different straight lines on the plot. 

<br/><br/>

Let's try to get a better intuition for this through the example below <br/>
Consider some training data (relation b/w size of house and it's price)

| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1.0               | 300                      |
| 2.0               | 500                      |

Let's plot a scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

# defining the styling of the plots
plt.style.use('util/deeplearning.mplstyle')

# using numpy to hold the training data
x_train = np.array([1.0, 2.0])         # features
y_train = np.array([300.0, 500.0])     # targets

# generate a scatter plot
plt.scatter(x_train, y_train, marker='x', color='red')
plt.title('House Price vs Size')
plt.ylabel('Price (1000s of dollars)')
plt.xlabel('Size  (1000 sqft)')
plt.show()
```

Now our goal is predict a line, that fit the training data.

Once we define the model to fit the training data, then with that model we can predict the price of houses of different sizes.

For that we have to identify the value of `w` and `b` that make the model fits the data.

Lets start with $w = 100$ and $b = 100$

```python
w = 100
b = 100


def compute_model_output(x_train: np.array, w :float, b :float) -> np.array:
    # size of training data
    m = x_train.shape[0] 

    # array to hold the predicted values
    f_wb = np.zeros(m)

    for i, x in enumerate(x_train):
        f_wb[i] = x * w + b

    return f_wb



# using the inital assumed w and b values, let's use the model to
# predict the house price
f_wb = compute_model_output(x_train, w, b)

# let's plot the predicted values in a plot along with scatter plot
plt.scatter(x_train, y_train, marker='x', color='red', label='training data')
plt.plot(x_train, f_wb, color='blue', label='Prediction')

plt.title('House Price vs Size')
plt.ylabel('Price (1000s of dollars)')
plt.xlabel('Size  (1000 sqft)')
plt.show()
```

with `w = 100` and `b = 100` the line don't fit out training data (red marker) <br/>

So, we have to experiment with `w` and `b` values that make the line fits the training data. <br/>
