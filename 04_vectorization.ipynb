#### NumPy - Scientific Computing Package

#### Vectors
- ordered collection of numbers
- elements of a vector are of same type
- denoted by lowercase bold letters $\mathbf{x}$ or by placing a arrow on top $\vec{x}$
- *dimension* or *rank* refers to size of a vector (no of values in the vector)

$$\vec{x} = \begin{pmatrix} x_1 \\ x_2 \\ x_3 \\ x_4 \\ x_n \end{pmatrix}$$

- this vector $\vec{x}$ has a dimension of `n`


#### VECTOR CREATION

```python
import numpy as np
import time

# allocates memory for vector and initialize with values
# input - dimension of the vector
a = np.zeros(4)
print(f'np.zeros(4): a = {a}, shape = {a.shape}, type = {a.dtype}')

b = np.zeros((4,4))
print(f'np.zeros(4,4): b = {b}, shape = {b.shape}, type = {b.dtype}')

c = np.random.random_sample(4)
print(f"np.random.random_sample(4): c = {c}, shape = {c.shape}, type = {c.dtype}")

# don't take dimension as input
d = np.arange(4.)
print(f'np.arange(4.): d = {d}, shape = {d.shape}, type = {d.dtype}')
```

#### Operations on Vectors

```python
e = np.array([1,2,3,4,5])

print(-e)
print(e ** 2)
print(np.sum(e))

f = -e
print(e + f)
```

#### Comparing Execution Speed of doing dot product using loop & using numpy

```python
np.random.seed(1)

# let's generate a huge data set
a = np.random.rand(1_00_00_000)
b = np.random.rand(1_00_00_000)

# let's calculate the time taken with loops
res = 0
start = time.time()

for i in range(a.shape[0]):
    res += a[i] * b[i]

end = time.time()
print(f'loop | result = {res}, time taken = {1000 * (end - start):.4f} ms')


# let's calculate the time taken with np.dot function
start = time.time()

res = np.dot(a, b)

end = time.time()
print(f'np.dot | result = {res}, time taken = {1000 * (end - start):.4f} ms')
```

Based on the consumed time it's clear that, vectorization provides a large speed up in this example. <br/>
This is because NumPy makes better use of available data parallelism in the underlying hardware. GPU's and modern CPU's implement Single Instruction, Multiple Data (SIMD) pipelines allowing multiple operations to be issued in parallel. This is critical in Machine Learning where the data sets are often very large.

#### MATRIX CREATION

```python
a = np.zeros((5, 3))  # matrix with 5 rows and 3 column
print(f'a = {a}, shape = {a.shape}')

b = np.zeros((1, 3))  # matrix with 1 rows and 3 column
print(f'\n\nb = {b}, shape = {b.shape}')

c = np.zeros((3, 1))  # matrix with 3 rows and 1 column
print(f'\n\nc = {c}, shape = {c.shape}')

d = np.zeros((1, 1))  # matrix with 1 rows and 1 column
print(f'\n\nd = {d}, shape = {d.shape}\n\n')


# arange - generates 1D array of numbers b/w 1 to 6
# 2  - specify we need a 2D matrix of column 2
# -1 - specify automatic row deterimination based on the no of values & column count
e = np.arange(6).reshape(-1, 2)
print(e, end="\n\n")

# in summary, (-1, 2) specify I need a matrix of two columns but automatically determine
# the no of rows based on the no of elements and column size

f = np.arange(6).reshape(2, -1)
print(f)
```

##### SLICING IN NUMPY

```python
a = np.arange(20).reshape(-1, 10)
print(a, end="\n\n")

slice1 = a[0, 1:5]
print(slice1, end="\n\n")

slice2 = a[:, 1:5]
print(slice2, end="\n\n")

slice3 = a[1, :]
print(slice3, end="\n\n")

print(a[1,4], end="\n\n")
```
