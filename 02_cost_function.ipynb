#### Cost Function
- purpose is to determines how well our line fits the training data

The equation for cost with one variable is:
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$ 
 
where 
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$
  
- $f_{w,b}(x^{(i)})$ is our prediction for example $i$ using parameters $w,b$.  
- $(f_{w,b}(x^{(i)}) -y^{(i)})^2$ is the squared difference between the target value and the prediction.   
- These differences are summed over all the $m$ examples and divided by `2m` to produce the cost, $J(w,b)$.

##### Why does the cost function reduce to get the best fit line?
As the line gets closer to the training data, the prediction errors become smaller. Since the cost function measures the total error, it also reduces. Therefore, minimizing the cost function naturally leads to finding the best fit line.

A contour plot is utilized to represent a 3D graph in a 2D format. It is generated by slicing the 3D surface and projecting each slice onto a 2D plane.

As we shift the point on the contour plot toward the origin, the corresponding point on the 3D graph moves down to the lowest point. This adjustment brings the line closer to the training data.
